```python
print(f'{'*'*100}')
print(f'{'*'*100}\n')
print('\t \t \t \t  Feature engineering')
print(f'\n{'*'*100}')
print(f'{'*'*100}')

print(f'\n{'*'*100}')
print('Sapertaing the Input feature  and target feature ')
print(f'{'*'*100}\n')

X=df_new.drop(columns=['y'])
y=df['y']

print(f'\n{'*'*100}')
print('Sapertaing the train and test data using train test split ')
print(f'{'*'*100}\n')

X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=42)


```

```python
ohe_columns=[col for col in X_train.columns if col !='education' and X_train[col].dtype =='object']
ohe_columns
```

```python
print(f'\n{'*'*100}')
print('Using columns transformer to transform categorical data')
print(f'{'*'*100}\n')

transformer=ColumnTransformer(
    transformers=(
        ('toe1',OrdinalEncoder(categories=[['university.degree','professional.course','high.school','basic.9y','basic.6y','basic.4y','illiterate','unknown']]),['education']),
        ('tohe2',OneHotEncoder(
            handle_unknown='ignore',
            sparse_output=False,
            drop='first'
        ),ohe_columns)
    ),remainder='passthrough'
)

X_train_new=transformer.fit_transform(X_train)
X_test_new=transformer.transform(X_test)
```
